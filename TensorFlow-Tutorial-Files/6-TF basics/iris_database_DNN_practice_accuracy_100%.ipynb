```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
feat_cols = [sepal_length, sepal_width, petal_length, petal_width]
```

```python
sepal_length = tf.feature_column.numeric_column('sepal_length')
sepal_width = tf.feature_column.numeric_column('sepal_width')
petal_length = tf.feature_column.numeric_column('petal_length')
petal_width = tf.feature_column.numeric_column('petal_width')
```

```python
#iris['species'] = iris['species'].map({'setosa': 1, 'versicolor': 2, 'virginia': 3})
iris['species'] = iris['species'].rank(method='dense', ascending=False).astype(int)
```

```python
df = iris.groupby('species')['sepal_width'].nunique()
```

```python
labels = iris['species']
```

```python
iris.head()
```

```python
x_data = iris.drop('species', axis=1)
```

```python
x_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size=0.3,random_state=101)
```

### tensor flow

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=20, num_epochs=5000, shuffle=True)
```

```python
model = tf.estimator.DNNClassifier(hidden_units=[10,20,20,10], feature_columns=feat_cols, n_classes=4)
```

```python
model.train(input_fn= input_func, steps=5000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, y=y_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
model.evaluate(eval_input_func)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x = X_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python

predictions = model.predict(pred_input_func)
```
